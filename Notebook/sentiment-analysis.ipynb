# Web-Scraping

```python
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.service import Service
from webdriver_manager.microsoft import EdgeChromiumDriverManager


def getReviews(id):
    
    driver = webdriver.Edge(service=Service(EdgeChromiumDriverManager().install()))

    url = f"https://www.themoviedb.org/movie/{id}/reviews"
    driver.get(url)

    
    # Extract the reviews
    reviews = driver.find_elements(By.CLASS_NAME, "teaser")  # Adjust class name to match the page's structure

    listOfReviews = []
    for review in reviews:
        listOfReviews.append(review.text)
        # print(review.text)


    
    driver.quit()

    return listOfReviews
```

```python
import requests
from bs4 import BeautifulSoup

# URL of the webpage to scrape

def getImg(currId):
    # Send a GET request to the URL
    url = f"https://www.themoviedb.org/movie/{currId}"

    response = requests.get(url)

    # Parse the HTML content with BeautifulSoup
    soup = BeautifulSoup(response.content, 'html.parser')

    # Find all img tags and extract their src attributes
    img_tags = soup.find_all('img',class_="poster w-full")
    for img in img_tags:
        src = img.get('src')  # Extract the src attribute
        print(src)
```

```python
file_path = r"../Data/reviews_rated.csv"
import csv

with open(file_path, mode='w', newline='') as file:
    writer = csv.DictWriter(file, fieldnames=["id", "review", "rating"])
    writer.writeheader()

    # Write the data rows using a for loop
    for id, review in all_reviews.iterrows():
        curr_rating  = pipe(review["review"])[0]["label"][0]
    
        writer.writerow({"id":review["id"],"review":str(review["review"]),"rating":int(curr_rating)})

print(f"Data has been written to {file_path}")
```
