# Builiding the recommender

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity



df = pd.DataFrame(filtered_data)

def get_id_from_name(name):
    
    return df['id'][df["original_title"]==name][0]


vectorizer = TfidfVectorizer(stop_words='english')

tfidf_matrix = vectorizer.fit_transform(df['overview'])

# Compute the cosine similarity between all movie plots
cosine_sim = cosine_similarity(tfidf_matrix, tfidf_matrix)

# Function to get the 5 most similar movies for a given movie title
def get_top_n_similar_movies(movie_id_num, n=5):
    
    movie_id = df[df['id'] == movie_id_num].index[0]
    similarities = cosine_sim[movie_id]
    
    similar_movie_ids = similarities.argsort()[-(n+1):-1]  # Exclude the movie itself
    
    # Get the movie titles of the most similar movies
    similar_movies = df['id'].iloc[similar_movie_ids].values


    return similar_movies


```

```python
similar_movies = get_top_n_similar_movies(879, n=5)
print(f"The 5 most similar movies are: {similar_movies}")
```

```python
similar_movies = get_top_n_similar_movies(9291, n=5)
print(f"The 5 most similar movies are: {similar_movies}")
```
